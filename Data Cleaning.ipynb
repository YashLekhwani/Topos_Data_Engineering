```python
import re
import numpy as np
import pandas as pd

dataset = pd.read_csv("dataset.csv")
```

```python
dataset
```

```python
dataset.rename(columns={'2016 land area': 'Total Area (sqmi)'}, inplace = True)
```

```python
dataset=dataset.fillna(0) #handle missing values
```

```python
dataset
```

```python
dataset['2018estimate'] = dataset['2018estimate'].str.replace(',', '')
dataset['2010Census'] = dataset['2010Census'].str.replace(',', '')
```

```python
dataset['Description'] = dataset['Description'].str.replace('\r', '')
dataset['Description'] = dataset['Description'].str.replace('\n', '')
```

```python
cleaning_list = ['+',"$","/",",","?",".",";","-","@","!","&","%","^","*",")","(", "\n"]
                                                                                
for item in cleaning_list:
    dataset['Description'] = dataset['Description'].str.replace(item," ")
```

```python
dataset
```

```python
dataset.to_csv('finaldataset.csv', index = True, header = True)
```
